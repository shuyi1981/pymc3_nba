```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
x = [i/2 for i in range(1,20)]
```

```python
print(x)
type(x)
```

```python
y = np.log(x)
y

```

```python
df =  pd.DataFrame({'y':y, 
'x':x,
'y2': y-max(y)})
```

```python
sns.stripplot(x = 'x', 
y = 'y2', 
data = df)
```

```python
sns.pointplot(x = 'x', 
y = 'y2', 
data = df)
```
